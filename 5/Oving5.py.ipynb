```python
import pandas as pd
data = pd.read_csv("agaricus-lepiota.csv")
data.head()
```

```python
data[['habitat', 'edibility']].groupby("habitat").describe()
#Count = antall sopper i hvert habitat
#Top viser mest vanlig sopp i habitatet e = edible(spiselig), p = poisonous(uspiselig)
#Freq sier hvor stor andel av habitatet som er spiselig/uspiselig
```

```python
dummies = pd.get_dummies(data)
dummies.head()
```

```python
import matplotlib.pyplot as plt 
plt.figure()
dummies[["edibility_e", "population_s"]].hist(color='k', alpha=1)
```
